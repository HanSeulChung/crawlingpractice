### 조문번호 및 조문내용 여러개

```python
bltags = soup.select('span.bl')
lawcontags = soup.select('div.lawcon')


number_list = []  #조문번호는 index1 =1조
content_list = []  #조문내용은 index10=1조문내용

maximum = len(lawcontags)

for i in range(maximum):
    if i+1<10:
        number_list.insert(i+1,bltags[i+1].text[1])
        content_list.insert(i,lawcontags[i].text)
    elif i+1>=10:
        number_list.insert(i+1,bltags[i+1].text[1:3])
        content_list.insert(i,lawcontags[i].text)

print(number_list)
print(content_list)
    
```

### url 하나에 있는 조문번호 및 조문내용 어러개 csv로 저장하기 전체 코드

```python
from selenium import webdriver

from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())


basic_link ="https://www.law.go.kr/"

url = basic_link+dfnew['법령상세링크2'][0]
driver.get(url)

html= driver.page_source

soup = BeautifulSoup(html,"html.parser")

bltags = soup.select('span.bl')
lawcontags = soup.select('div.lawcon')

number_list = []  #조문번호는 index1 =1조
content_list = []  #조문내용은 index10=1조문내용

maximum = len(lawcontags)

for i in range(maximum):
    if i+1<10:
        number_list.insert(i+1,bltags[i+1].text[1])
        content_list.insert(i,lawcontags[i].text)
    elif i+1>=10:
        number_list.insert(i+1,bltags[i+1].text[1:3])
        content_list.insert(i,lawcontags[i].text)
        
```
